````python
import matplotlib as mpl
import numpy as np
````

## **`pyplot`**

Matplotlib has a number of sub-libraries. The most important one is `pyplot`. Since it deserves a special attention, so we will go through some of its important functions.






````python
import matplotlib.pyplot as plt
````

### Figure Creation

We can create a new figure as:

`<figure name> = plt.figure()`

````python
fig1 = plt.figure()
````

Lets take a few mins to understand its technical details: A figure in pyplot uses:

- **Inches** as its height, width units, not cms.
- A default dpi (dots per inch) of **100.0**.
- Default dimensions of $ 6.4 \times 4.8 \space in^2$. Hence, we are seeing `640x480` pixels as above figure's dimensions.

We can override the default values as per requirement. For example:

````python
fig1 = plt.figure(figsize=(12.80,9.60))
````

We can even set a higher dpi (and other attributes).

````python
fig1.dpi = 200
fig1
````

### Setting Axes

The figures _drawn_ so far are empty. We need to add a couple of other elements as well to complete them.

#### `plt.axes()`

For setting axes, we use `plt.axes()`.

````python
axis1 = plt.axes()
````

We can also override it to have the rectangle's dimensions (not coordinates) specified explicitly as:

```
<rectangle> = ((left, bottom, width, height))
<axis> = plt.axes(<rectangle>)
```

````python
axis2 = plt.axes((0,0,0.8,0.8))
````

#### `add_axes()`
There is no point in having an axis without an associated figure.

The correct way to do it to make the figure, followed by adding the axes. For that, we can use **`add_axes()`**. Its syntax is:

`<axis> = <figure variable>.add_axes(<rectangle tuple>)`

**Important:** There is no need to override the axes (i.e, use this function) if we are unsure whether we need it.

````python
ax11 = fig1.add_axes((0.5,0.5,1,1))
fig1
````

#### Visualizing `rect` parameters

We specified above that the rectangle takes a tuple:

$$(left, bottom, width, height)$$

We can see it in action here by adding several axes inside the same figure.

````python
fig3 = plt.figure()

axis1 = fig3.add_axes((0.5,0.5,0.5,0.5))

fig3
````

````python
axis2 = fig3.add_axes((1,1,0.5,0.5))


fig3
````

````python
axis3 = fig3.add_axes((1.5,1.5,0.5,0.5))

fig3
````

````python
plt.close()
````

### Plotting the Data

Here comes the most important part. i.e, plotting the data. We have set up the figure's dimensions, axes, dpi, etc. but yet to plot the actual data. For that, we will use the `plot()` method.

`<axes>.plot(<x>,<y>)`

**Note:** Since it requires axes, so we can save them in a suitable variable while using `add_axes()`.

````python
fig4 = plt.figure(dpi=200) #Just to remind that we can override the parameters.

axis4 = fig4.add_axes((0,0,0.5,0.5))

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

axis4.plot(x,y)
````

#### Setting the legend

We have made the graph. Good, but there's more to it. For example, setting the legend.

We can set up the legend in two steps:

1. Setting the `label` attribute in `plot()`.
2. Calling the `legend()` method.

````python
fig4 = plt.figure(dpi=200) #Just to remind that we can override the parameters.

axis4 = fig4.add_axes((0,0,0.5,0.5))

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

axis4.plot(x,y,label="Quadratic Regression")
axis4.legend()
````
